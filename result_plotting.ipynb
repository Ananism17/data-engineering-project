```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cores = [2, 4, 6, 8]
time_cores = [30.96, 30.05, 29.84, 30.57]  # Execution time for core scaling

memory = [1, 2, 4, 8]  # Memory in GB
time_memory = [30.05, 30.25, 31.6, 30.4]  # Execution time for memory scaling

partitions = [50, 100, 200, 500]
time_partitions = [29.38, 28.91, 28.79, 29.25]  # Execution time for shuffle partitions

data_size = [100000, 300000, 500000]  # Number of songs
data_time = [28.91, 76.82, 125.72]  # Execution time for dataset scaling
speed = [d/t for d, t in zip(data_size, data_time)]  # Data processed per second
```

```python
# Figure 1: CPU Core Scaling
plt.figure(figsize=(10, 6))
plt.plot(cores, time_cores, marker='o', linestyle='-', color='b', label='Execution Time')
plt.xlabel("Number of Cores")
plt.ylabel("Execution Time (s)")
plt.title("Effect of CPU Core Scaling on Execution Time")
plt.legend()
plt.grid()
plt.savefig("figure1_core_scaling.png")
```

```python
# Figure 2: Memory Scaling
plt.figure(figsize=(10, 6))
plt.plot(memory, time_memory, marker='s', linestyle='-', color='g', label='Execution Time')
plt.xlabel("Memory per Executor (GB)")
plt.ylabel("Execution Time (s)")
plt.title("Effect of Memory Scaling on Execution Time")
plt.legend()
plt.grid()
plt.savefig("figure2_memory_scaling.png")
```

```python
# Figure 3: Shuffle Partition Tuning
plt.figure(figsize=(10, 6))
plt.plot(partitions, time_partitions, marker='^', linestyle='-', color='r', label='Execution Time')
plt.xlabel("Number of Shuffle Partitions")
plt.ylabel("Execution Time (s)")
plt.title("Effect of Shuffle Partition Tuning on Execution Time")
plt.legend()
plt.grid()
plt.savefig("figure3_shuffle_partitions.png")
```

```python
# Figure 4: Dataset Scaling
plt.figure(figsize=(10, 6))
plt.plot(data_size, speed, marker='d', linestyle='-', color='purple', label='Data Processed per Second')
plt.xlabel("Dataset Size (Number of Songs)")
plt.ylabel("Data Processed per Second")
plt.title("Effect of Dataset Scaling on Processing Speed")
plt.legend()
plt.grid()
plt.savefig("figure4_dataset_scaling.png")
```
